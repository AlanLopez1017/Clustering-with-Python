# Elbow method and silhouette coefficient

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn import metrics
from scipy.spatial.distance import cdist
from sklearn.metrics import silhouette_samples, silhouette_score
```

```python
x1 = np.array([3,1,1,2,1,6,6,6,5,6,7,8,9,8,9,9,8])
x2 = np.array([5,4,5,6,5,8,6,7,6,7,1,2,1,2,3,2,3])
X = np.array(list(zip(x1,x2))).reshape(len(x1),2)
```

```python
plt.xlim([0,10])
plt.ylim([0,10])
plt.xlabel("x")
plt.ylabel("y")
plt.title("Dataset to classify")
plt.scatter(x1,x2)
plt.show()
```

```python
max_k = 10 # maximum number of clusters that we are going to create
K = range(1, max_k+1)
ssw = [] # sum of internal squares
color_palette = [plt.cm.Spectral(float(i)/max_k) for i in K]
centroid = [sum(X)/len(X) for _ in K] # general centroid
sst = sum(np.min(cdist(X, centroid, "euclidean"), axis = 1))

for k in K:
    kmeanModel = KMeans(n_clusters = k).fit(X)
    centers = pd.DataFrame(kmeanModel.cluster_centers_)
    labels = kmeanModel.labels_

    ssw_k = sum(np.min(cdist(X, kmeanModel.cluster_centers_, "euclidean"), axis = 1))
    ssw.append(ssw_k)

    label_color = [color_palette[i] for i in labels]

    ## Making a silhouette for each cluster
    # For security reasons we do not make a silhouette if k = 1 or k = len(X)
    if 1 < k < len(X):
        ## Create a subplot with one row and two columns (silhouette and cluster segmentation)
        fig, (axis1, axis2) = plt.subplots(1,2)
        fig.set_size_inches(20,8)

        # The first subplot will contain the silhouette which can have values from -1 to 1.
        # In this case we already control that the values are between -0.1 and 1
        axis1.set_xlim([-0.1,1.0])
        # The number of clusters to be inserted will determine the size of each bar
        # The coefficient (n_clusters + 1)*10 will be the blank space that will be left between individual silhouettes of each cluster to separate them
        axis1.set_ylim([0,len(X)+(k+1)*10])

        silhouette_avg = silhouette_score(X, labels)
        print("*  For k = {} the average of the silhouette is: ".format(k, silhouette_avg))
        sample_silhouette_values = silhouette_samples(X, labels)

        y_lower = 10

        for i in range(k):
            # We add the silhouette of the ith-cluster
            ith_cluster_sv = sample_silhouette_values[labels == i]
            print("     - For i = {} the silhouette of the cluster is: {}".format(i+1, np.mean(ith_cluster_sv)))
            # We arrange the silhouettes of the ith cluster in descending order
            ith_cluster_sv.sort()

            # We calculate where to place the first silhouette on the vertical axis
            ith_cluster_size = ith_cluster_sv.shape[0]
            y_upper = y_lower + ith_cluster_size

            # We choose the color of the cluster
            color = color_palette[i]

            # We paint the silhouette of the i-th cluster
            axis1.fill_betweenx(np.arange(y_lower, y_upper),
                                0, ith_cluster_sv, facecolor = color, alpha = 0.7)
            

            # We label said cluster with the number in the center
            axis1.text(-0.05, y_lower + 0.5*ith_cluster_size, str(i+1))

            # We calculate the new y_lower for the next cluster in the graph
            y_lower = y_upper + 10 # We leave 10 positions empty without sample

        axis1.set_title("Silhouette representation for k = %s"%str(k))
        axis1.set_xlabel("S(i)")
        axis1.set_ylabel("Cluster ID")
    
        # End of silhouette representation

    # K-means plot with respective points
    plt.plot()
    plt.xlim([0,10])
    plt.ylim([0,10])
    plt.title("Clustering for k = %s"%str(k))
    plt.scatter(x1,x2, c = label_color)
    plt.scatter(centers[0], centers[1], marker="x")
    plt.show()


```

```python
# Representacin del codo
plt.plot(K, ssw, "bx-")
plt.xlabel("k")
plt.ylabel("SSw(k)")
plt.title("Elbow technique to find the optimal k")
plt.show()
```

We see that the elbow is at k = 3, which coincides with the k = 3 graph where the silhouettes were more homogeneous (presented less discrepancy with the value) of the global silhouette of the entire cluster (0.7122). So it seems that the silhouette and elbow technique coincide.

Using the normalized elbow technique:

```python
# Representacin del codo normalizado
plt.plot(K, 1-ssw/sst, "bx-")
plt.xlabel("k")
plt.ylabel("1-nomr(SSw(k))")
plt.title("Normalized elbow technique to find the optimal k")
plt.show()
```

At k = 3 it stops believing and remains more linear and it doesn't make much sense to continue doing more divisions.
